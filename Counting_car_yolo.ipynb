```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
!pip install supervision
import supervision as sv
import numpy as np
import cv2
```

```python
!mkdir custom_test_data
```

```python
video_path = "/content/drive/MyDrive/Colab Notebooks/counting_car_yolo_proj/custom_test_data/highway_car2.mp4"
output_path = "/content/drive/MyDrive/Colab Notebooks/counting_car_yolo_proj/highway_car_counted4.mp4"
```

```python
video_info = sv.VideoInfo.from_video_path(video_path)
print("Video size:", video_info.width, video_info.height)
```

```python
line_zone = sv.LineZone(
    start=sv.Point(x=191, y=319),
    end=sv.Point(x=1279, y=319)
)
```

```python
box_annotator = sv.BoxAnnotator()
line_annotator = sv.LineZoneAnnotator(thickness=6, text_thickness=4, text_scale=2.0)
```

```python
with sv.VideoSink(output_path, video_info) as sink:
    for result in model.track(
        source=video_path,
        stream=True,
        persist=True,
        classes=[2],
        imgsz=640,
        tracker="bytetrack.yaml"
    ):
        frame = result.orig_img
        detections = sv.Detections.from_ultralytics(result)

        if result.boxes.id is not None:
            detections.tracker_id = result.boxes.id.cpu().numpy().astype(int)

        line_zone.trigger(detections)

        frame = box_annotator.annotate(frame, detections=detections)
        frame = line_annotator.annotate(frame, line_zone)

        cv2.putText(frame, f"Cars out: {line_zone.out_count}", (50, 80),
                    cv2.FONT_HERSHEY_SIMPLEX, 2.5, (0, 255, 0), 3)

        sink.write_frame(frame)
```
